```python
import numpy as np
import matplotlib.pyplot as plt
file_path = 'raw-training-log.txt'
data_raw = np.loadtxt(file_path, skiprows=2, unpack=True)
file_path = 'ours-training-log.txt'
data_ours = np.loadtxt(file_path, skiprows=2, unpack=True)
file_path = '4-block-ours-training-log.txt'
data_ours_4blks = np.loadtxt(file_path, skiprows=2, unpack=True)
#file_path = '8-block-ours-training-log.txt'
#data_ours_8blks = np.loadtxt(file_path, skiprows=2, unpack=True)
```

```python
print(f'Raw best Val Loss: {np.min(data_raw[2])}')
print(f'Our 2-blocks best Val Loss: {np.min(data_ours[2])}')
print(f'Our 4-blocks best Val Loss: {np.min(data_ours_4blks[2])}')
# print(f'Our 8-blocks best Val Loss: {np.min(data_ours_8blks[2])}')
```

```python
#plt.plot(data[0], data[1], marker='.', linestyle='-', color='b', label='MSE Loss')
plt.plot(data_raw[0], data_raw[1], linestyle='-', label='Raw-TL')
plt.plot(data_raw[0], data_raw[2], linestyle='-', label='Raw-VL')
plt.plot(data_ours[0], data_ours[1], linestyle='-',label='Ours-2blks-TL')
plt.plot(data_ours[0], data_ours[2], linestyle='-', label='Ours-2blks-TL')
plt.plot(data_ours_4blks[0], data_ours_4blks[1], linestyle='-',label='Ours-4blks-TL')
plt.plot(data_ours_4blks[0], data_ours_4blks[2], linestyle='-', label='Ours-4blks-TL')
#plt.plot(data_ours_8blks[0], data_ours_8blks[1], linestyle='-',label='Ours-8blks-TL')
#plt.plot(data_ours_8blks[0], data_ours_8blks[2], linestyle='-', label='Ours-8blks-TL')
plt.title("Loss Curve")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.legend()
plt.show()
plt.savefig('Loss_curve.png') 
```
